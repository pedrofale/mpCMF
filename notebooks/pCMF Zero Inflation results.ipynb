```python
%matplotlib inline

from pCMF.misc import plot_utils

import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
sns.set_style('whitegrid')
```

# Silhouette scores

```python
silh_pca_scores = np.load('../scripts/simulations/zero_inflation/output/silh_pca_scores.npy')
silh_gap_scores = np.load('../scripts/simulations/zero_inflation/output/silh_gap_scores.npy')
silh_zigap_scores = np.load('../scripts/simulations/zero_inflation/output/silh_zigap_scores.npy')
```

```python
# Prepare arrays to plot
no_zi_low_sep = [silh_pca_scores[0, 0, :], silh_gap_scores[0, 0, :], silh_zigap_scores[0, 0, :]]
no_zi_high_sep = [silh_pca_scores[0, 1, :], silh_gap_scores[0, 1, :], silh_zigap_scores[0, 1, :]]

low_zi_low_sep = [silh_pca_scores[1, 0, :], silh_gap_scores[1, 0, :], silh_zigap_scores[1, 0, :]]
low_zi_high_sep = [silh_pca_scores[1, 1, :], silh_gap_scores[1, 1, :], silh_zigap_scores[1, 1, :]]

high_zi_low_sep = [silh_pca_scores[2, 0, :], silh_gap_scores[2, 0, :], silh_zigap_scores[2, 0, :]]
high_zi_high_sep = [silh_pca_scores[2, 1, :], silh_gap_scores[2, 1, :], silh_zigap_scores[2, 1, :]]
```

```python
# Plot silhouette scores
legend=['PCA', 'GaP', 'ZIGaP']
fig = plt.figure(figsize=(15, 4))

ax1 = plt.subplot(1, 3, 1)
plot_utils.plot_simulation_results(no_zi_low_sep, no_zi_high_sep, ax=ax1, ylabel='Silhouette score', title='No ZI')

ax2 = plt.subplot(1, 3, 2, sharey=ax1)
plot_utils.plot_simulation_results(low_zi_low_sep, low_zi_high_sep, ax=ax2, title='Low ZI')

ax3 = plt.subplot(1, 3, 3, sharey=ax1)
_, handles = plot_utils.plot_simulation_results(high_zi_low_sep, high_zi_high_sep, ax=ax3, legend=legend, title='High ZI')

plt.legend(handles, legend, labelspacing=0.5, bbox_to_anchor=(1.3, 0.9))
_ = [h.set_visible(False) for h in handles]

plt.show()
```
